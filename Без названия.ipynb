```python
import os
import requests
import django
os.environ.setdefault('DJANGO_SETTINGS_MODULE', 'blog.settings')
os.environ.setdefault('DJANGO_ALLOW_ASYNC_UNSAFE', "true")
django.setup()
```

```python
from django.contrib.auth import get_user_model
User = get_user_model()
user = User.objects.get(id=1)
```

```python
from article.models import Post, Comment, Like
```

```python
Post.objects.all().delete()
```

```python
url = "https://jsonplaceholder.typicode.com/posts"
response = requests.get(url)
data = response.json()
post_cid_to_id_dict = {}
for post in data:
    p = Post.objects.create(title=post['title'],slug=post['id'],body=post['body'], author=user)
    post_cid_to_id_dict[post['id']] = p.id
```

```python
Post.objects.all().delete()
```

```python
url = "https://jsonplaceholder.typicode.com/comments"
data = requests.get(url).json()
for comment in data:
    post = Post.objects.get(slug=comment['postId'])
    Comment.objects.create(post=post, body=comment['body'], author=comment['email'])
```
